```python
import turtle
from turtle import *
from turtle import hideturtle
from tkinter import *
from PIL import Image
```

```python
base = 50
sample_n = 100
step_n = 350 / sample_n
file_n = 1
Circle_size = 1
Square_size = 1
Triangle_size = 1
cord = (base/2)
```

```python
def square(base):  
    for side in range(4):
        t.fd(base)
        t.rt(90)

def triangle(base):
    t.fd((base/2))
    for side in range(3):
        t.rt(120)
        t.fd(base)
        
def circle(base):
    t.circle(base)

def save_square(file_n, Square_size):
    image_eps = ("Archive/Square_size-"+str(Square_size)+"-N-"+str(file_n)+".eps")
    im = Image.open(image_eps)
    fig = im.convert('L')
    image_png = ("Shapes/Square/Square_size-"+str(Square_size)+"-N-"+str(file_n)+".png")
    fig.save(image_png, lossless=True)

def save_triangle(file_n, triangle_size):
    image_eps = ("Archive/Triangle_size-"+str(triangle_size)+"-N-"+str(file_n)+".eps")
    im = Image.open(image_eps)
    fig = im.convert('L')
    image_png = ("Shapes/Triangle/Triangle_size-"+str(triangle_size)+"-N-"+str(file_n)+".png")
    fig.save(image_png, lossless=True)
    
def save_circle(file_n, Circle_size):
    image_eps = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    im = Image.open(image_eps)
    fig = im.convert('L')
    image_png = ("Shapes/Circle/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".png")
    fig.save(image_png, lossless=True)
```

```python
s = turtle.Screen()
s.screensize(24,24)
t = turtle.Turtle()
t.hideturtle()
t.pensize(4)
```

```python
for j in range(sample_n):
    t.home()         #        SQUARE
    t.goto(0,(-cord-(cord/2.5)))
    t.clear()
    t.lt(135)
    square(base)    #   DIAMOND
    Square_shape = A
    file = ("Archive/Square_size-"+str(Square_shape)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_square(file_n, Square_size)   
    Square_size = Square_size +1

    t.home()
    t.goto(-cord,-cord)
    t.clear()
    t.lt(90)
    square(base)    #    STRAIGHT
    file = ("Archive/Square_size-"+str(Square_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_square(file_n, Square_size)
    Square_size = Square_size +1

    t.home()           #        TRIANGLE
    t.goto(0,-cord)
    t.clear()
    t.lt(180)
    triangle(base)    #    tri upper
    Triangle_des = 
    file = ("Archive/Triangle_size-"+str(Triangle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_triangle(file_n, Triangle_size)
    Triangle_des = Triangle_size +1
    
    t.home()
    t.goto(0,cord)
    t.clear()
    triangle(base)    #     tri lower
    file = ("Archive/Triangle_size-"+str(Triangle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_triangle(file_n, Triangle_size)
    Triangle_size = Triangle_size +1
    
    t.home()
    t.goto(cord, 0)
    t.clear()
    t.rt(90)
    triangle(base)    #    tri left
    file = ("Archive/Triangle_size-"+str(Triangle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_triangle(file_n, Triangle_size)
    Triangle_size = Triangle_size +1
    
    t.home()
    t.goto(-cord, 0)
    t.clear()
    t.lt(90)
    triangle(base)    #     tri right
    file = ("Archive/Triangle_size-"+str(Triangle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_triangle(file_n, Triangle_size)
    Triangle_size = Triangle_size +1
    
    t.home()           #       CIRCLE
    t.goto(0,-cord)
    t.clear()
    circle(base/2)     #     CENTER
    file = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_circle(file_n, Circle_size)
    Circle_size = Circle_size +1
    
    t.home()
    t.goto(-cord,-cord)
    t.clear()
    t.rt(45)
    circle(base/2)     #      bottom left
    file = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_circle(file_n, Circle_size)
    Circle_size = Circle_size +1
    
    t.home()
    t.goto(cord, -cord)
    t.clear()
    t.lt(45)
    circle(base/2)     #     bottom right
    file = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_circle(file_n, Circle_size)
    Circle_size = Circle_size +1
    
    t.home()
    t.goto(cord, cord)
    t.clear()
    t.lt(135)
    circle(base/2)     #     top right
    file = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_circle(file_n, Circle_size)
    Circle_size = Circle_size +1
    
    t.home()
    t.goto(-cord, cord)
    t.clear()
    t.rt(135)
    circle(base/2)      #     top left
    file = ("Archive/Circle_size-"+str(Circle_size)+"-N-"+str(file_n)+".eps")
    s.getcanvas().postscript(file=file)
    save_circle(file_n, Circle_size)
    Circle_size = Circle_size +1

    file_n = file_n + 1
    Circle_size = 10
    Square_size = 10
    Triangle_size = 10
    
    base = base + step_n
    cord = cord + (step_n/2)
```
